```r
observations <- read.csv("../observations.csv")
head(observations)
```

```r
mean_temp <- mean(observations$Température, na.rm = TRUE)
var_temp  <- var(observations$Température, na.rm = TRUE)
print(mean_temp)
print(var_temp)
```

```r
# t.test(observations$Température, mu = 288.56)
t.test(observations$Température, mu = mean_temp)
```

```r
hiver <- subset(observations, mois_de_l_annee %in% c(12,1,2))$Température
ete   <- subset(observations, mois_de_l_annee %in% c(6,7,8))$Température
t.test(hiver, ete)
```

```r
tab1 <- table(observations$mois_de_l_annee, observations$Type.de.tendance.barométrique)
chisq.test(tab1)
```

```r
anova_res <- aov(Température ~ factor(mois_de_l_annee), data = observations)
summary(anova_res)
```
