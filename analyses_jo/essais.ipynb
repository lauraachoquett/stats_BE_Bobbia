```r
# Sélectionner uniquement les colonnes numériques météo
cols_utiles <- c("Température", "Humidité", "Point.de.rosée", "Pression.au.niveau.mer",
                 "Vitesse.du.vent.moyen.10.mn", "Latitude", "Longitude", "Altitude")
meteo_small <- data_meteo[, cols_utiles]

summary(meteo_small)
```

```r
# 2. Présentation graphique
library(ggplot2)


# Histogramme de la température (K)
ggplot(meteo_small, aes(x=Température)) + geom_histogram(bins=50, fill="skyblue", color="black") +
ggtitle("Distribution des températures")


# Nuage de points Température vs Humidité
ggplot(meteo_small, aes(x=Température, y=Humidité)) + geom_point(alpha=0.5) +
ggtitle("Température vs Humidité relative")

# Nuage de points Température vs Point de rosée
ggplot(meteo_small, aes(x=Température, y=Point.de.rosée)) + geom_point(alpha=0.5) +
ggtitle("Température vs Point de rosée")
```

```r
# Nuage de points Température vs Point de rosée
ggplot(meteo_small, aes(x=Température, y=Altitude)) + geom_point(alpha=0.5) +
ggtitle("Température vs Altitude")
```

```r
# Regrouper par mois ou saisons

library(lubridate)
data_observ$Date <- as.POSIXct(data_meteo$Date, format="%Y-%m-%dT%H:%M:%S", tz="Europe/Berlin")
data_observ$mois <- month(data_meteo$Date)
data_observ$saison <- ifelse(data_meteo$mois %in% c(12, 1, 2), "Hiver",
                            ifelse(data_meteo$mois %in% c(3, 4, 5), "Printemps",
                                   ifelse(data_meteo$mois %in% c(6, 7, 8), "Été", "Automne")))
```

```r
library(dplyr)

# Nombre d'observations par saison
saison_counts <- data_observ %>%
    group_by(saison) %>%
    summarize(nb_obs = n())
print(saison_counts)

ggplot(saison_counts, aes(x = saison, y = nb_obs, fill = saison)) +
  geom_col() +
  labs(title = "Répartition des observations par saison",
       x = "Saison", y = "Nombre d'observations") +
  theme_minimal()
```
