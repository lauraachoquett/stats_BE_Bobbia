```r
# Sélectionner uniquement les colonnes numériques météo
cols_utiles <- c("Température", "Humidité", "Point.de.rosée", "Pression.au.niveau.mer",
                 "Vitesse.du.vent.moyen.10.mn", "Latitude", "Longitude", "Altitude")
meteo_small <- data_meteo[, cols_utiles]

summary(meteo_small)
```

```r
# 2. Présentation graphique
library(ggplot2)


# Histogramme de la température (K)
ggplot(meteo_small, aes(x=Température)) + geom_histogram(bins=50, fill="skyblue", color="black") +
ggtitle("Distribution des températures")


# Nuage de points Température vs Humidité
ggplot(meteo_small, aes(x=Température, y=Humidité)) + geom_point(alpha=0.5) +
ggtitle("Température vs Humidité relative")
```

```r
# Estimateurs température

mean_t <- mean(meteo_small$`Température`, na.rm=TRUE)
median_t <- median(meteo_small$`Température`, na.rm=TRUE)
sd_t <- sd(meteo_small$`Température`, na.rm=TRUE)
mean_t -273.15
mean_t
median_t
sd_t
```

```r
# Regrouper par mois ou saisons

library(lubridate)
data_meteo$Date <- as.POSIXct(data_meteo$Date, format="%Y-%m-%dT%H:%M:%S", tz="Europe/Berlin")
data_meteo$mois <- month(data_meteo$Date)
data_meteo$saison <- ifelse(data_meteo$mois %in% c(12, 1, 2), "Hiver",
                            ifelse(data_meteo$mois %in% c(3, 4, 5), "Printemps",
                                   ifelse(data_meteo$mois %in% c(6, 7, 8), "Été", "Automne")))
```
